```python
# plot the annotation file and image path and plot the boxes on img
import cv2
import matplotlib.pyplot as plt
import numpy as np

def plot_dota_annotation(image_path, annotation_path):
	# Load the image
	image = cv2.imread(image_path)

	# Read the DOTA annotations
	with open(annotation_path, "r") as file:
		lines = file.readlines()

	# Plot each bounding box
	for line in lines:
		values = line.strip().split()
		coordinates = list(map(float, values[:8]))
		class_name = values[8]

		# Draw the bounding box
		pts = np.array(coordinates, np.int32).reshape((-1, 1, 2))
		cv2.polylines(image, [pts], isClosed=True, color=(0, 255, 0), thickness=1)

		# Draw the class name
		# cv2.putText(image, class_name, (coordinates[0], coordinates[1]), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

	# Display the image
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.figure(figsize=(10, 10))
	plt.imshow(image)
	plt.axis("off")
	plt.show()

plot_dota_annotation("./random/test_img.jpg", "./random/test_ann.txt")
```
